## Project description 

`The main idea of this project is to predict the dependent variable (coordinates of the middle point of the box containing the mouse), based on a set of independent variables (coordinates of 7 joints of the mouse and the coordinates of the 2-point containing box). This is done by designing an automated model based on an artificial neural network. The training dataset is generated from the given images by annotations a set of points for each mouse in the images.`

### 2. **Initialize dataset function.**

This function arranges the data and stores it in a data frame, the purpose of this is for ease of analysis and dealing with the data.



```python
def Initialize_dataset_func(df,columns_name):
    data=pd.DataFrame(columns=columns_name)
    mouses=[df[(df.columns.levels[0][0],df.columns.levels[1][i])]  for i in range(len(df.columns.levels[1]))]
    for mouse in mouses:
        for index,img in enumerate(list(mouse.index.droplevel([0,1]))):
            row={'image_name':img}
            for point in mouse.iloc[index].index:
                name=point[0]+'_'+point[1]
                row[name]=mouse.iloc[index][point[0]][point[1]]

            data=data.append(row, ignore_index=True)

    return data

#............................................
```

```python
columns_name=['image_name','topleft_x','topleft_y','rightdown_x','rightdown_y',
             'nose_x','nose_y','leftear_x','leftear_y','rightear_x','rightear_y'
             ,'leftHip_x','leftHip_y','rightHip_x','rightHip_y','tailBase_x',
              'tailBase_y','tailEnd_x','tailEnd_y']

#............................................

data=Initialize_dataset_func(df,columns_name)
data
```

### 3. **Clean data function.**

This function cleans the data by deleting all records that contain at least one empty point.
Since it is not logical to use methods for processing null values in the data, for example, we cannot use average points. Therefore, in order to build an accurate model, it is better to delete these records.



```python
def Clean_data_func(data):
    data_clean=data.loc[~np.isnan(data['topleft_x'])]
    data_clean=data_clean.dropna(axis=0)
    
    return data_clean
    
#............................................
```

### 4. **Show image function.**

This function displays an image and draws points and boxes on it, in order to observe and know the data more.

```python
def Show_image_func(root_path,image_name,data,Pcolor='red',Rcolor='green'):
    image=cv2.imread(root_path+image_name)
    image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
   
    dataf=data.loc[data['image_name']==image_name]
    if len(dataf)>2:
        dataf=dataf.loc[0:2]
    
    for indexf in range(dataf.shape[0]):
        x_points=dataf.iloc[indexf].values[1::2]
        y_points=dataf.iloc[indexf].values[2::2]
        left, bottom, width, height = (x_points[0], y_points[0], x_points[1]-x_points[0], y_points[1]-y_points[0])
        rect=mpatches.Rectangle((left,bottom),width,height,fill=False,color=Rcolor,linewidth=1.5)
        plt.gca().add_patch(rect)
        plt.scatter(x_points, y_points,c =Pcolor,s=10)
    
    
#............................................
        
```

### 5. **initialize target variable (dependent variable) functions.**

The dependent variable, which will be predicted by the machine learning model to be designed, represents the midpoint of the mouse enclosure. Since the dependent variable is not present in the data set, the dependent variable (the midpoint of the box) was formed through the two points $(topleft,rightdown)$ as in the following equation:
\begin{equation} 
\label{eq:1}
\begin{split}
center_x&=\frac{topleft_x+rightdown_x}{2}\\
\\
center_y&=\frac{topleft_y+rightdown_y}{2}\\
\end{split}
\end{equation}

```python
def Initialize_target_variable(data):
    datanew=copy.deepcopy(data)
    center_x=(data.topleft_x.values+data.rightdown_x.values)/2
    center_y=(data.topleft_y.values+data.rightdown_y.values)/2
    datanew['center_x']=center_x
    datanew['center_y']=center_y
    return datanew

#............................................
```

```python
data_end=Initialize_target_variable(data_cleand)
data_end
```

### 6. **Functions for augment the data.**

After deleting the NaN rows, our total number of records decreased from 250 to 67. So we need to create new images to increase the number of data. Since our work depends on a set of points only, therefore we do not need to produce new images, but we need to produce new points i.e. mouse New), and in order to increase the number of rats I simply applied geometric transformations to the rat points in the data set, by applying the rotation transform by angles $(30,60,90,120,150,180,210,240,270,300)$ to the points of each mouse in the data set, I used homogeneous matrices to perform the rotation transformation As follows.

\begin{equation} 
\label{eq:2}
\begin{split}
P^{(org)}& =\begin{bmatrix}
        x_1 & y_1 & 1 \\
        x_2 & y_2 & 1 \\
        \vdots & \vdots & \vdots \\
        x_{10} & y_{10} & 1 \\
        \end{bmatrix}\\ \\
T & =\begin{bmatrix}
        1 & 0 & 0 \\
        0 & 1 & 0 \\
        -x_c & -y_c & 1 \\
        \end{bmatrix} * \begin{bmatrix}
                        \cos\theta & \sin\theta & 0 \\
                         -\sin\theta & \cos\theta & 0 \\
                        0 & 0 & 1 \\
        \end{bmatrix}* \begin{bmatrix}
                        1 & 0 & 0 \\
                        0 & 1 & 0 \\
                        x_c & y_c & 1 \\
        \end{bmatrix}\\ \\
        = &\begin{bmatrix}
                \cos\theta & \sin\theta & 0 \\
                -\sin\theta & \cos\theta & 0 \\
                -x_c\cos\theta+y_c\sin\theta+x_c &-x_c\sin\theta-y_c\cos\theta+y_c & 0 \\
        \end{bmatrix}
\end{split}
\end{equation}

\begin{equation} 
\label{eq:3}
P^{(new)}=P^{(org)}*T
\end{equation}

The previous rotation matrix is applied to each mouse in the data set at different angles. This means that from each mouse in the original dataset, 10 mice with different orientations will be produced.

In the end, we obtained 737 mouse (record) as our final data set which we will use to build a neural network model.


```python
def Rotation_func(xpoints,ypoints,anlge,width=640,height=360):
    xc=width/2
    yc=height/2
    cos=np.cos(np.deg2rad(anlge))
    sin=np.sin(np.deg2rad(anlge))
    
    p_mtrix=np.ones((len(xpoints),3))
    p_mtrix[:,0]=xpoints
    p_mtrix[:,1]=ypoints
    
    T_mtrix=np.array([[cos,sin,0],[-sin,cos,0],
                      [-xc*cos+yc*sin+xc,-xc*sin-yc*cos+yc,0]
                     ])
    
    pn_mtrix=np.dot(p_mtrix,T_mtrix)
    
    return pn_mtrix[:,0],pn_mtrix[:,1]
    
    
#............................................

def Augment_data_func(data,anlges,width=640,height=360):
    data_augm=copy.deepcopy(data)
    for anlge in anlges:
        for index in range(data.shape[0]):
            row=data.iloc[index]
            x_new,y_new=Rotation_func(row.values[1::2],row.values[2::2],anlge,width,height)
            row[1::2]=x_new
            row[2::2]=y_new
            data_augm=data_augm.append(row,ignore_index=True)
            
        
    return data_augm
        
#............................................




```
